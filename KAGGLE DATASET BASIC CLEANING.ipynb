## SUMMARY OF THE DATA
### THIS CONTAINS IMPORTANT DATA OF AUDIOBOOKS RELEASED FROM 1998 TO 2025
IT CONTAINS 87489 ROWS AND 8 COLUMNS

## COLUMN DETAILS
### It contains 9 columns
- **NAME**: Name of audiobook
- **AUTHOR**: Author of the audiobook
- **NARRATOR**: Narrator of the audiobook
- **TIME**: Length of the audiobook
- **RELEASEDATE**: Release date of the audiobook
- **LANGUAGE**: Language of the audiobook
- **STARS**: Number of stars the audiobook received
- **PRICE**: Price of the audiobook in INR
- **RATINGS**: Number of reviews received by the audiobook

# Issues Identified During Manual Inspection

## Dirty Data Issues

1. Column names are not standardized and need to be corrected.

2. The `Name` column contains:
   - Spelling mistakes
   - Character encoding issues (e.g., `â€™`, `Ã©`, `Ð`, `Ñ`)

3. Remove the prefix `WrittenBy:` from the `Written By` column.

4. Author names should be separated into:
   - First Name
   - Last Name

5. The `Written By` column contains encoding issues.

6. Remove the prefix `NarratedBy:` from the `Narrator` column.

7. The `Narrator` column contains encoding issues.

8. Narrator names should be separated into:
   - First Name
   - Last Name

9. The `Time` column should be converted to an appropriate datetime/time format.

10. Many records contain `Anonymous` as the author or narrator also some contain uncredited replace them with unknown

11. The `Release Date` column should be converted to datetime format.
    - Mixed date formats are present:
      - `DD-MM-YYYY`
      - `MM-DD-YYYY`

12. The `Language` column contains inconsistent capitalization:
    - Uppercase values
    - Lowercase values
    - Title Case values

13. The `Price` column should be converted to a numeric data type also some entries are free which we need to change to 0

14. The `Rating` column contains some empty values which are represented as 'Not rated yet"

## UPON AUTOMATIC ASSESSMENT FOLLOWING ISSUES WERE FOUND
1. Some entries in author do not end with an alphabet.
2. Some entries in narrator do not end with an alphabet.
3. Some duration entries contain "Less than 1 min".
4. Some author entries contain role keywords such as editor, illustrator, traductor, translator, etc.

---

## Untidy Data Issues

1. The `Ratings` column contains multiple variables in a single column:
   - Star Rating
   - Number of Reviews

   These should be separated into two independent columns:

   - `Stars`
   - `Number_of_Reviews`
